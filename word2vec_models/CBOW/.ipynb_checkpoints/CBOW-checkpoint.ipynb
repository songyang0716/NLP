### Data Input

```python
size = 10000
review_json = pd.read_json(yelp_dataset_dir, 
                      lines=True,
                      chunksize=size)

review_df = pd.DataFrame()

# on include the first 100k reviews
for review_chunck in review_json:
    review_df = pd.concat([review_df, review_chunck])
    break

nlp = en_core_web_sm.load()
```

```python
reviews = review_df['text'].values
```

```python
with open("yelp_review_10000.txt", 'w') as output:
    for row in reviews:
        output.write(str(row) + '\n')
```

```python
len(reviews)
```

```python
CONTEXT_SIZE = 2  # 2 words to the left, 2 to the right
EMBEDDING_DIM = 50
vocab = set()
contexts = []


for index, review in enumerate(reviews):
    if index % 1000 == 0:
        print("Processing index: ", index)
    doc = nlp(review)
    for sentence in doc.sents:
        clean_sentence = [word.text.lower() for word in sentence if (not (word.is_stop) and (word.is_alpha))]
        for i in range(2, len(clean_sentence)-2):
            context = [clean_sentence[i - 2], clean_sentence[i - 1],
                       clean_sentence[i + 1], clean_sentence[i + 2]]
            target = clean_sentence[i]
            contexts.append((context, target))
    
```

```python
for context, target in contexts:
    vocab.add(target)
    vocab.add(context[0])
    vocab.add(context[1])
    vocab.add(context[2])
    vocab.add(context[3])
```

```python
len(vocab)
```

### CBOW Model

```python
vocab_size = len(vocab)
word_to_ix = {word: i for i, word in enumerate(vocab)}
ix_to_word = {i: word for i, word in enumerate(vocab)}
```

```python
class CBOW(nn.Module):
    def __init__(self, vocab_size, embedding_dim, context_size):
        super(CBOW, self).__init__()
        self.embeddings = nn.Embedding(vocab_size, embedding_dim)
        self.linear = nn.Linear(embedding_dim, vocab_size)

    def forward(self, inputs):
        embeds = self.embeddings(inputs)
        embeds_mean = torch.mean(embeds, dim=1)
        score = self.linear(embeds_mean)
        log_probs = F.log_softmax(score, dim=1)
        return log_probs 
```

```python
losses = []
loss_function = nn.NLLLoss()
model = CBOW(len(vocab), EMBEDDING_DIM, CONTEXT_SIZE)
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
epoches = 100
BATCH_SIZE = 100
```

```python
for epoch in range(epoches):
    print(epoch)
    total_loss = 0
    n_batch = int(len(contexts) / BATCH_SIZE)
    for i_batch in range(n_batch):
        cur_batch_context = contexts[i_batch*BATCH_SIZE: (i_batch+1)*BATCH_SIZE]
        context_idxs = torch.tensor([[word_to_ix[w] for w in context[0]] for context in cur_batch_context], dtype=torch.long)
        log_probs = model(context_idxs)
        loss = loss_function(log_probs, torch.tensor([word_to_ix[context[1]]  for context in cur_batch_context], dtype=torch.long))
        model.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    losses.append(total_loss)
print(losses)
```
